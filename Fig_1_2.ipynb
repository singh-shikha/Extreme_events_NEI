```python
#making an empty array
array1 = [[] for _ in range(61)]
#adding all the no. of events when rainfall is more than 50mm for the year 1960 to 2019
for i in range(1960,1990):
                    startdate = datetime(year=i, month= 1 , day= 1)
                    enddate = datetime(year=i, month= 12, day=31 )
                    startdate,enddate
                    da = ds11.sel(TIME=slice(startdate, enddate),LATITUDE=slice('26','30'),LONGITUDE=slice('94','98'))
                    a=da.RAINFALL
                    b=np.where(a > 50, 1, 0)
                    np.max(b),np.min(b)
                    y = np.nansum(b,axis= 1)
                    z = np.nansum(y,axis= 1)
                    p=np.nansum(z)
                    array1=np.append(array1,p)

for i in range(1990,2020):
                    startdate = datetime(year=i, month= 1 , day= 1)
                    enddate = datetime(year=i, month= 12, day=31 )
                    startdate,enddate
                    da = ds21.sel(TIME=slice(startdate, enddate),LATITUDE=slice('26','30'),LONGITUDE=slice('94','98'))
                    a=da.RAINFALL
                    b=np.where(a > 50, 1, 0)
                    np.max(b),np.min(b)
                    y = np.nansum(b,axis= 1)
                    z = np.nansum(y,axis= 1)
                    p=np.nansum(z)
                    array1=np.append(array1,p)
                    
x = np.arange(1960, 2020)
from scipy.stats import linregress
linregress(x,array1) #x and y are arrays or lists.

```
