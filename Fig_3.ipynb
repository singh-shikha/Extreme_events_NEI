```python
#no of events with rainfall greater than 50mm for years 1960 to 1987
a=ds1.p
b=np.where(a > 50, 1, 0)
np.max(b),np.min(b)
X1=np.nansum(b,axis=0)
X1.shape


a=ds2.p
b=np.where(a > 50, 1, 0)
np.max(b),np.min(b)
Y1=np.nansum(b,axis=0)
Y1.shape
```

```python
class MidpointNormalize(mpl.colors.Normalize):
    """Normalise the colorbar."""
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        mpl.colors.Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [-200, 0,200]
        return np.ma.masked_array(np.interp(value, x, y), np.isnan(value))

divnorm=colors.TwoSlopeNorm(vmin=-200., vcenter=0., vmax=200)
mp.contourf(x,y,(X1-Y1), norm=divnorm,cmap='seismic',levels=np.arange(-200,200))
```
